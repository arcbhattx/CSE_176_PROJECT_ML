```python
import pandas as pd

df = pd.read_csv("./datasets/social_media_vs_productivity.csv")
```

```python
# Unique values for object datatypes
# To be turned into real values for model
df_columns = df.columns

for column in df_columns:
    if df[column].dtype == "object" or df[column].dtype == "bool":
        print(df[column].unique())
```

```python
# turns all unique values to a number through mapping

for column in df_columns:
    if df[column].dtype == "object" or df[column].dtype == "bool":
        unique_values = df[column].unique()
        mapping = {value: i for i, value in enumerate(unique_values)}
        df[column] = df[column].map(mapping)
            
```

```python
df.head()
```

```python
# replace NaN values with 0

df.fillna(0, inplace=True)
```
